```python
import pandas as pd
```

```python
def checkOverlap(row):
    oo,ot = int(row.one.split("-")[0]),int(row.one.split("-")[1])
    to,tt = int(row.two.split("-")[0]),int(row.two.split("-")[1])
    if (oo <= to) and (ot >= tt):
        return 1
    if (to <= oo) and (tt >= ot):
        return 1
    return 0
```

```python
df = pd.read_csv("data/day4", sep =",", skip_blank_lines=True,header=None)
df.columns = ["one","two"] 
df["overlaps"] = df.apply(lambda row: checkOverlap(row), axis = 1)
print("Overlaps: ",df.overlaps.sum())
df
```

```python
def checkCommon(row):
    oo,ot = int(row.one.split("-")[0]),int(row.one.split("-")[1])
    to,tt = int(row.two.split("-")[0]),int(row.two.split("-")[1])
    # Checks si oneone plus grand que twoone et plus petit que twotwo
    if (oo >= to) and (oo <= tt):
        return 1
    if (ot >= to) and (ot <= tt):
        return 1
    if (to >= oo) and (to <= ot):
        return 1
    if (tt >= oo) and (tt <= ot):
        return 1
    return 0
```

```python
df["Common"] = df.apply(lambda row: checkCommon(row), axis = 1)
print("Common: ",df.Common.sum())
# 930 too high
df[df.Common == 0]
```
