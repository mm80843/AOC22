```python
def cleanPath(path,ix):
    if not path.startswith("/"):
        path = "/"+path
    lst = path.split("/")
    index = [k for k in range(len(lst)) if lst[k] == ".."]
    RES = []
    for item in lst:
        if item == "..":
            try:
                RES.pop()
            except:
                print(ix," -- error with",path)
        else:
            RES.append(item)
    #cleaned = "/".join(lst).replace("///","/").replace("///","/")
    #cleaned = cleaned.replace("//","/")
    cleaned = "/".join(RES)
    return cleaned
```

```python
DF = df.groupby("cleanPath").SIZE.sum().reset_index()
DF
```

```python
def cumulSize(row):
    ttlSize = 0
    miniDF = DF[DF.cleanPath.str.startswith(row.cleanPath)]
    ttlSize = miniDF.SIZE.sum()
    return ttlSize
```

```python
DF["ttlSize"] = DF.apply(lambda x:cumulSize(x), axis=1)
DF.sort_values(by="cleanPath")
```
